```python
import nltk
nltk.download('punkt')

import fitz  # PyMuPDF
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

# Download 'punkt' resource
import nltk
nltk.download('punkt')

file_path = '/content/Story of Prophets.pdf'  # Replace with the actual path to your PDF file

story = []

with fitz.open(file_path) as pdf_document:
    for page_number in range(pdf_document.page_count):
        page = pdf_document[page_number]
        text = page.get_text()
        raw_sent = sent_tokenize(text)
        for sent in raw_sent:
            story.append(simple_preprocess(sent))

```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('muhammad')
```

```python
model.wv.doesnt_match(['yusuf','yahya','muhammad','harun','who'])
```
