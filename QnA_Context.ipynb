```python
import torch
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
```

```python
QA_input = [{
    'question': 'Why is model conversion important?',
    'context': 'The option to convert models between FARM and transformers gives freedom to the user and let people easily switch between frameworks.'},
    {
    'question': 'What type of large language model is GPT-4?',
    'context': 'Generative Pre-trained Transformer 4 (GPT-4) is a multimodal large language model created by OpenAI, and the fourth in its series of GPT foundation models. It was launched on March 14, 2023, and made publicly available via the paid chatbot product ChatGPT Plus, and via OpenAI API.'}
            ]
```

```python
model_name = 'deepset/roberta-base-squad2'
```

```python
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
inputs0 = tokenizer(QA_input[0]['question'], QA_input[0]['context'],return_tensors="pt")
outputs0 = model(**inputs0)
```

```python
inputs1 = tokenizer(QA_input[1]['question'], QA_input[1]['context'],return_tensors="pt")
outputs1 = model(**inputs1)
```

```python
outputs1
```

```python
answer_start_idx = torch.argmax(outputs1.start_logits)
answer_end_idx = torch.argmax(outputs1.end_logits)

answer_tokens = inputs1.input_ids[0, answer_start_idx: answer_end_idx + 1]
answer = tokenizer.decode(answer_tokens)
print("ques: {}\nanswer: {}".format(QA_input[1]['question'], answer))
```
